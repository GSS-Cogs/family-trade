```python
from databaker.framework import *
```

```python
import pandas as pd 
```

```python
inputFile = 'balanceofpayments2017q3.xls'
```

```python
tab = loadxlstabs(inputFile)
```

```python
tab = tab[3]
```

```python
savepreviewhtml(tab)
```

```python
observations = tab.excel_ref('D9').expand(DOWN).expand(RIGHT).is_not_blank()
```

```python
savepreviewhtml(observations)
```

```python
Code = tab.excel_ref('C9').expand(DOWN).is_not_blank()
```

```python
savepreviewhtml(Code)
```

```python
Year = tab.excel_ref('D4').expand(RIGHT).is_not_blank()
Quarter = tab.excel_ref('D5').expand(RIGHT)
```

```python
savepreviewhtml(Year)
```

```python
savepreviewhtml(Quarter)
```

```python
Account = tab.excel_ref('B').expand(DOWN).by_index([7,27,47,67]) - tab.excel_ref('B78').expand(DOWN)
```

```python
savepreviewhtml(Account)
```

```python
Services = tab.excel_ref('B').expand(DOWN).by_index([8,13,19,28,33,39,48,53,59,68,25,45,65,76])
```

```python
savepreviewhtml(Services)
```

```python
Nettransactions = tab.excel_ref('B').is_not_blank().is_not_whitespace() - Account - Services
```

```python
Nettransactions = Nettransactions - tab.excel_ref('B78').expand(DOWN) -tab.excel_ref('B2').expand(UP)
```

```python
savepreviewhtml(Nettransactions)
```

```python
Currency = tab.excel_ref('N3')
```

```python
savepreviewhtml(Currency)
```

```python
Dimensions = [
            HDimConst('GEOG', 'K02000001'),
            HDim(Year,'Year',DIRECTLY,ABOVE),
            HDim(Quarter,'Quarter',DIRECTLY,ABOVE),
            HDim(Code,'Code',DIRECTLY,LEFT),
            HDimConst('Currency', 'GBP(millions)'),            
            HDim(Nettransactions,'Type of Services',DIRECTLY,LEFT),
            HDim(Services,'Services',CLOSEST, ABOVE),
            HDim(Account,'Account Transactions',CLOSEST,ABOVE)
            
    
]
```

```python
c1 = ConversionSegment(observations, Dimensions, processTIMEUNIT=True)
savepreviewhtml(c1)
```

```python
new_table = c1.topandas()
new_table
```

```python
new_table['TIME'] = new_table['Year'].map(lambda cell:cell.replace('.0', ''))
new_table
```

```python
print(len(new_table))
new_table['TIME'] = new_table['TIME'].map(str) + new_table['Quarter'] 
```

```python
print(len(new_table))
new_table
```

```python
new_table = new_table.drop('Year', axis=1)
new_table = new_table.drop('Quarter', axis=1)
```

```python
print(len(new_table))
new_table
```

```python
new_table['Account Transactions'] = new_table['Account Transactions'].map(lambda x: x.rstrip('1')) 
```

```python
new_table.tail(5)
```

```python

new_table.fillna('', inplace=True)

writetechnicalCSV('balanceofpayments2017q3tabB.csv', new_table)
```
