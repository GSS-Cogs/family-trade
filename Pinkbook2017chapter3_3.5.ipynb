```python
tab = loadxlstabs(pbFile, sheetids='3.5')[0]
savepreviewhtml(tab)
```

```python
observations = tab.excel_ref('D8').expand(DOWN).expand(RIGHT).is_not_blank()
```

```python
Code = tab.excel_ref('C8').expand(DOWN).is_not_blank()
```

```python
Year = tab.excel_ref('D3').expand(RIGHT).is_not_blank()
```

```python
Commerce = tab.excel_ref('B').expand(DOWN).by_index([5,39,59]) - tab.excel_ref('B72').expand(DOWN)
```

```python
Financialservices = tab.excel_ref('B').expand(DOWN).by_index([7,28,37,41,50,57,61,68,70,71]) - tab.excel_ref('B72').expand(DOWN)
```

```python
Revenue = tab.excel_ref('B').expand(DOWN).is_not_blank().is_not_whitespace() - Commerce - Financialservices - tab.excel_ref('B1').expand(UP)
```

```python
Currency = tab.excel_ref('W2')
```

```python
Dimensions = [
            HDimConst('GEOG', 'K02000001'),
            HDim(Year,'TIME', DIRECTLY,ABOVE),
            HDim(Code,'Cdid',DIRECTLY,LEFT),
            HDimConst('Currency', 'Pounds(millions)'),            
            HDim(Revenue, 'Revenue', DIRECTLY, LEFT),
            HDim(Financialservices, 'Financial Services', CLOSEST, ABOVE),
            HDim(Commerce,'Commerce', CLOSEST, ABOVE)    
]
```

```python
c1 = ConversionSegment(observations, Dimensions, processTIMEUNIT=True)
```

```python
savepreviewhtml(c1)
```

```python
new_table = c1.topandas()
new_table
```

```python
new_table = new_table.drop('TIMEUNIT', axis=1)
```

```python
print(len(new_table))
new_table
```

```python
new_table['Financial Services'] = new_table['Financial Services'].map(lambda x: x.rstrip('123'))
```

```python
new_table
```

```python
new_table['Revenue'] = new_table['Revenue'].fillna('')
```

```python
new_table['Revenue'] = new_table['Revenue'].map(lambda x: x.rstrip('23'))
```

```python
new_table
```

```python
new_table['TIME'] = new_table['TIME'].str.strip()
new_table['GEOG'] = new_table['GEOG'].str.strip()
new_table['Currency'] = new_table['Currency'].str.strip()
new_table['Cdid'] = new_table['Cdid'].str.strip()
```

```python
new_table['Commerce'] = new_table['Commerce'].str.strip()
```

```python
new_table['Revenue'] = new_table['Revenue'].str.strip()
```

```python
new_table['Financial Services'] = new_table['Financial Services'].str.strip()
```

```python
destinationFolder = Path('out')
destinationFolder.mkdir(exist_ok=True)

pbCSV = destinationFolder / (Path(pbFilename).stem + '_3.5.csv')

new_table.to_csv(pbCSV, index = False)
```
