```python
from databaker.framework import *
import pandas as pd 
from io import BytesIO
```

```python
import requests
from cachecontrol import CacheControl
from cachecontrol.caches.file_cache import FileCache
from cachecontrol.heuristics import LastModified
from pathlib import Path

session = CacheControl(requests.Session(),
                       cache=FileCache('.cache'),
                       heuristic=LastModified())

sourceFolder = Path('in')
sourceFolder.mkdir(exist_ok=True)

inputURL = 'https://www.uktradeinfo.com/Statistics/Tax%20and%20Duty%20Bulletins/Alcohol0718.xls'
```

```python
data = BytesIO(session.get(inputURL).content)
```

```python
tab = pd.read_excel(data, sheet_name = '10')
```

```python
data = temp.iloc[4:]
```

```python
data.columns.values[0] = 'Year'
data.columns.values[1] = 'UK Beer Production'
data.columns.values[2] = 'UK Alcohol Production'
data.columns.values[6] = 'Alcohol Clearances'
data.columns.values[7] = 'Cider Clearances'
```

```python
new_table = pd.melt(data, id_vars=['Year'], var_name='Category', value_name='Value')
```

```python
new_table
```

```python
new_table.count()
```

```python
new_table.dropna(how='any',axis=0, inplace =True)
```

```python
new_table.count()
```

```python
new_table.dtypes
```

```python
def user_perc(x):
    
    if '*'in str(x)  :
        return 'estimated based on previous years'
    else:
        return ''
    
new_table['Revision'] = new_table.apply(lambda row: user_perc(row['Value']), axis = 1)
```

```python
new_table['Value'] = new_table['Value'].astype(str).map(lambda cell: cell.replace('*', ''))
```

```python
new_table['Value'] = pd.to_numeric(new_table['Value'], errors='coerce')
```

```python
new_table.count()
```

```python
new_table['Category'].unique()
```

```python
new_table['Alcohol Content'] = 'Various'
```

```python
def user_perc5(x):
    
    if ((str(x) == 'Total Beer'))| ((str(x) == 'Total Cider')) | ((str(x) == 'Total Alcohol')): 
        
        return '£ million'
    else:
        return 'Thousand hectolitres'
    
new_table['Unit'] = new_table.apply(lambda row: user_perc5(row['Category']), axis = 1)
```

```python
def user_perc6(x):
    
    if ((str(x) == 'Total Beer'))| ((str(x) == 'Total Cider')) | ((str(x) == 'Total Alcohol')) : 
        
        return 'GBP Total'
    else:
        return 'Net Volume'
    
new_table['Measure Type'] = new_table.apply(lambda row: user_perc6(row['Category']), axis = 1)
```

```python
def user_perc7(x,y):
    
    if ((str(x) ==  '2018-05-01 00:00:00')) | ((str(x) == '2018-06-01 00:00:00')) | ((str(x) == '2018-07-01 00:00:00')): 
        
        return 'provisional'
    else:
        return y
    
new_table['Revision'] = new_table.apply(lambda row: user_perc7(row['Year'],row['Revision']), axis = 1)
```

```python
def user_perc8(x,y):
    
    if ((str(x) == 'Total Alcohol')): 
        
        return ''
    else:
        return y
    
new_table['Alcohol Content'] = new_table.apply(lambda row: user_perc8(row['Category'], row['Alcohol Content']), axis = 1)
```

```python
new_table = new_table[['Year','Category','Alcohol Content','Measure Type','Value','Unit','Revision']]
```

```python
destinationFolder = Path('out')
destinationFolder.mkdir(exist_ok=True, parents=True)

new_table.to_csv(destinationFolder / ('hmrcalcohol_10.csv'), index = False)
```
