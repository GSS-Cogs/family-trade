```python
tab = loadxlstabs('PB2017_3.4.xls')
```

```python
tab = tab[0]
savepreviewhtml(tab)
```

```python
observations = tab.excel_ref('D9').expand(DOWN).expand(RIGHT).is_not_blank()
```

```python
Code = tab.excel_ref('C9').expand(DOWN).is_not_blank()
```

```python
Year = tab.excel_ref('D4').expand(RIGHT).is_not_blank()
```

```python
Commerce = tab.excel_ref('B').expand(DOWN).by_index([6,21,30]) - tab.excel_ref('B38').expand(DOWN)
```

```python
InsuranceRevenue = tab.excel_ref('B').expand(DOWN).is_not_blank().is_not_whitespace() - Commerce - tab.excel_ref('B2').expand(UP)
```

```python
Currency = tab.excel_ref('W3')
```

```python
Dimensions = [
            HDimConst('GEOG', 'K02000001'),
            HDim(Year,'Year', DIRECTLY,ABOVE),
            HDim(Code,'Code Identifier',DIRECTLY,LEFT),
            HDimConst('Currency', 'GBP(millions)'),            
            HDim(InsuranceRevenue, 'Insurance Revenue', DIRECTLY, LEFT),
            HDim(Commerce, 'Commerce', CLOSEST, ABOVE)    
]
```

```python
c1 = ConversionSegment(observations, Dimensions, processTIMEUNIT=True)
```

```python
savepreviewhtml(c1)
```

```python
new_table = c1.topandas()
new_table
```

```python
new_table = new_table.drop('DATAMARKER', axis=1)
```

```python
new_table['TIME'] = new_table['Year'].map(lambda cell:cell.replace('.0', ''))
new_table
```

```python
new_table = new_table.drop('Year', axis=1)
```

```python
print(len(new_table))
new_table
```

```python
new_table['Insurance Revenue'] = new_table['Insurance Revenue'].map(lambda x: x.rstrip('123'))
```
