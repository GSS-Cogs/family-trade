###  Balanceofpayments2017q3, TabC

```python
from databaker.framework import *
```

```python
import pandas as pd 
```

```python
inputFile = 'balanceofpayments2017q3.xls'
```

```python
tab = loadxlstabs(inputFile)
```

```python
tab = tab[4]
```

```python
savepreviewhtml(tab)
```

```python
observations = tab.excel_ref('D10').expand(DOWN).expand(RIGHT).is_not_blank()
```

```python
savepreviewhtml(observations)
```

```python
Code = tab.excel_ref('C10').expand(DOWN).is_not_blank()
```

```python
savepreviewhtml(Code)
```

```python
Year = tab.excel_ref('D4').expand(RIGHT).is_not_blank()
Quarter = tab.excel_ref('D5').expand(RIGHT)
```

```python
savepreviewhtml(Year)
```

```python
savepreviewhtml(Quarter)
```

```python
Region = tab.excel_ref('B').expand(DOWN).by_index([7,42]) - tab.excel_ref('B77').expand(DOWN)
```

```python
savepreviewhtml(Region)
```

```python
Transactions = tab.excel_ref('B').expand(DOWN).by_index([9,20,31,44,55,66])
```

```python
savepreviewhtml(Transactions)
```

```python
Services = tab.excel_ref('B').is_not_blank().is_not_whitespace() - Region - Transactions
```

```python
Services = Services - tab.excel_ref('B77').expand(DOWN) -tab.excel_ref('B2').expand(UP)
```

```python
savepreviewhtml(Services)
```

```python
Currency = tab.excel_ref('N3')
```

```python
savepreviewhtml(Currency)
```

```python
Dimensions = [
            HDimConst('GEOG', 'K02000001'),
            HDim(Year,'Year',DIRECTLY,ABOVE),
            HDim(Quarter,'Quarter',DIRECTLY,ABOVE),
            HDim(Code,'Code Identifier',DIRECTLY,LEFT),
            HDimConst('Currency', 'GBP(millions)'),            
            HDim(Services,'Services',CLOSEST, ABOVE),
            HDim(Transactions,'Type of Transactions',CLOSEST,ABOVE),
            HDim(Region,'Transactions Region',CLOSEST,ABOVE)
            
    
]
```

```python
c1 = ConversionSegment(observations, Dimensions, processTIMEUNIT=True)
savepreviewhtml(c1)
```

```python
new_table = c1.topandas()
new_table
```

```python
new_table['TIME'] = new_table['Year'].map(lambda cell:cell.replace('.0', ''))
new_table
```

```python
print(len(new_table))
new_table['TIME'] = new_table['TIME'].map(str) + new_table['Quarter'] 
```

```python
print(len(new_table))
new_table
```

```python
new_table = new_table.drop('Year', axis=1)
new_table = new_table.drop('Quarter', axis=1)
```

```python
print(len(new_table))
new_table
```

```python
new_table['Transactions Region'] = new_table['Transactions Region'].map(lambda x: x.rstrip('123456,')) 
```

```python
new_table.tail(5)
```

```python

new_table.fillna('', inplace=True)

writetechnicalCSV('balanceofpayments2017q3tabc.csv', new_table)
```
