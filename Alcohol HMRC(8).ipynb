```python
new_table.dropna(how='any',axis=0, inplace =True)
```

```python
new_table['Value'] = pd.to_numeric(new_table['Value'], errors='coerce')
```

```python
new_table.count()
```

```python
new_table['Category'].unique()
```

```python
new_table['Category'] = new_table['Category'].map(
    lambda x: {
        'Above 1.2% but not exceeding 5.5% ABV ' : 'Above 1.2% but not exceeding 5.5% ABV',
        'Imported \nex-ship' : 'Imported ex-ship',
        'Total wine1 ': 'Total Wine'
        }.get(x, x))
```

```python
def user_perc2(x):
    
    if ((str(x) ==  'Still')) | ((str(x) == 'Sparkling')): 
        
        return 'Above 5.5% ABV but not exceeding 15%'
    else:
        return 'Composition by Origin above 5.5% ABV'
    
new_table['Alcohol Content'] = new_table.apply(lambda row: user_perc2(row['Category']), axis = 1)
```

```python
def user_perc3(x,y):    
    if x.strip() == 'Above 1.2% but not exceeding 5.5% ABV':
        return 'Above 1.2% but not exceeding 5.5% ABV'
    else:
        return y        
    
new_table['Alcohol Content'] = new_table.apply(lambda row: user_perc3(row['Category'], row['Alcohol Content']), axis = 1)
```

```python
def user_perc4(x,y):
    
    if ((str(x) ==  'Total made wine')) | ((str(x) == 'Total Wine'))| ((str(x) == 'Total Alcohol')): 
        
        return 'Total'
    else:
        return y
    
new_table['Alcohol Content'] = new_table.apply(lambda row: user_perc4(row['Category'], row['Alcohol Content']), axis = 1)
```

```python
def user_perc5(x):
    
    if ((str(x) == 'Total Wine'))| ((str(x) == 'Total Alcohol')): 
        
        return '£ million'
    else:
        return 'hectolitres'
    
new_table['Unit'] = new_table.apply(lambda row: user_perc5(row['Category']), axis = 1)
```

```python
def user_perc6(x):
    
    if ((str(x) == 'Total Wine'))| ((str(x) == 'Total Alcohol')): 
        
        return 'GBP Total'
    else:
        return 'Net Volume'
    
new_table['Measure Type'] = new_table.apply(lambda row: user_perc6(row['Category']), axis = 1)
```

```python
def user_perc7(x):
    
    if ((str(x) ==  '2018-05-01 00:00:00')) | ((str(x) == '2018-06-01 00:00:00')) | ((str(x) == '2018-07-01 00:00:00')): 
        
        return 'provisional'
    else:
        return ''
    
new_table['Revision'] = new_table.apply(lambda row: user_perc7(row['Year']), axis = 1)
```

```python
new_table = new_table[['Year','Category','Alcohol Content','Measure Type','Value','Unit','Revision']]
```
